```python




# dataset Overview
print("Shape (rows, columns):", data.shape)
print("\nColumn Names:\n", data.columns.tolist())
print("\nData Types:\n", data.dtypes)

print("\nMissing Values per Column:\n", data.isnull().sum())
print("\nBasic Statistics:\n", data.describe(include='all'))  # includes both numeric & categorical


# Check how many duplicate rows exist
print("Number of duplicate rows:", data.duplicated().sum())

#  View duplicate rows (if any)
print("\nDuplicate rows:")
print(data[data.duplicated()])



```

```python


# ðŸ”¹ Separate columns by type
cat_cols = data.select_dtypes(include=['object', 'category']).columns
num_cols = data.select_dtypes(include=['number']).columns
date_cols = data.select_dtypes(include=['datetime64[ns]']).columns

print("Categorical Columns:", cat_cols.tolist())
print("Numerical Columns:", num_cols.tolist())
print("Datetime Columns:", date_cols.tolist())

# ðŸ”¹ Insights for categorical columns
print("\n=== Categorical Column Insights ===")
for col in cat_cols:
    missing_pct = data[col].isnull().mean() * 100
    print(f"\nColumn: {col}")
    print("Missing %:", round(missing_pct, 2))
    print("Unique values:", data[col].nunique())
    if data[col].nunique() > 0:
        print("Most frequent value:", data[col].mode()[0])
    print("Top 5 value counts:\n", data[col].value_counts().head())

# ðŸ”¹ Insights for numerical columns
print("\n=== Numerical Column Insights ===")
for col in num_cols:
    missing_pct = data[col].isnull().mean() * 100
    print(f"\nColumn: {col}")
    print("Missing %:", round(missing_pct, 2))
    print(data[col].describe())

# ðŸ”¹ Insights for datetime columns
print("\n=== Datetime Column Insights ===")
for col in date_cols:
    missing_pct = data[col].isnull().mean() * 100
    print(f"\nColumn: {col}")
    print("Missing %:", round(missing_pct, 2))
    if data[col].notnull().sum() > 0:
        print("Min date:", data[col].min())
        print("Max date:", data[col].max())
        print("Unique dates:", data[col].nunique())
```

```python


print("=== Potential Data Quality Issues ===")



# 3. Constant columns (only 1 unique value â†’ no useful info)
constant_cols = [col for col in data.columns if data[col].nunique() <= 1]
print("\n Constant Columns:", constant_cols)

# 4. High cardinality categorical columns
cat_cols = data.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    unique_ratio = data[col].nunique() / len(data)
    if unique_ratio > 0.9:  # >90% unique values = suspicious for categorical
        print(f"\n High cardinality column (possible ID): {col} ({data[col].nunique()} unique values)")

# 5. Outliers in numeric columns (using IQR rule)
num_cols = data.select_dtypes(include=['number']).columns
for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[col] < (Q1 - 1.5 * IQR)) | (data[col] > (Q3 + 1.5 * IQR))]
    if not outliers.empty:
        print(f"\n Possible outliers detected in column: {col} ({len(outliers)} rows)")

# 6. Datetime columns with strange ranges
date_cols = data.select_dtypes(include=['datetime64[ns]']).columns
for col in date_cols:
    if data[col].notnull().sum() > 0:
        min_date, max_date = data[col].min(), data[col].max()
        print(f"\nðŸ”¹ Datetime column {col}: min={min_date}, max={max_date}")
```
